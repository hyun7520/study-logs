```python
def solution(park, routes):
    answer = []

    xLimit = len(park) - 1
    yLimit = len(park[0]) - 1

    for i in range(len(park)):
        for j in range(len(park[1])):
            if park[i][j] == 'S':
                x, y = i, j

    tempX = x
    tempY = y

    for route in routes:
        moveLen = int(route[2])
        if route[0] == 'E': 
            if (tempY + moveLen) > yLimit:
                continue
            for i in range(1, moveLen + 1):
                tempY += 1
                if park[x][tempY] == 'X':
                    tempY = y
                    break

        elif route[0] == 'S':
            if (tempX + moveLen) > xLimit:
                continue
            for i in range(1, moveLen + 1):
                tempX += 1
                if park[tempX][y] == 'X':
                    tempX = x
                    break

        elif route[0] == 'W':
            if (tempY - moveLen) < 0:
                continue
            for i in range(1, moveLen + 1):
                tempY -= 1
                if park[x][tempY] == 'X':
                    tempY = y
                    break
  
        else:
            if (tempX - moveLen) < 0:
                continue
            for i in range(1, moveLen + 1):
                tempX -= 1
                if park[tempX][y] == 'X':
                    tempX = x
                    break

        x = tempX
        y = tempY

    answer.extend([x, y])

    return answer
```
